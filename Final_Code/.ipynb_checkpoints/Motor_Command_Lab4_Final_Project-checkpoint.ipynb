# 1. Set up the notebook

Import modules.

```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from IPython.display import display, IFrame, HTML
```

Define a function to load flight data from hardware experiments and resample it at 100 Hz using linear interpolation, optionally discarding data from the start or end of the experiment.

```python
def load_hardware_data(filename, t_min_offset=0, t_max_offset=0):
    # load raw data
    with open(filename, 'r') as f:
        data = json.load(f)

    # convert lists to numpy arrays
    for val in data.values():
        for key in val.keys():
            val[key] = np.array(val[key])

    # create an array of times at which to subsample
    t_min = -np.inf
    t_max = np.inf
    for key, val in data.items():
        t_min = max(t_min, val['time'][0])
        t_max = min(t_max, val['time'][-1])
    t_min += t_min_offset * 1000
    t_max -= t_max_offset * 1000
    nt = int(1 + np.floor((t_max - t_min) / 10.))
    t = np.arange(0, 10 * nt, 10) / 1000.
    resampled_data = {'time': t}

    # resample raw data with linear interpolation
    for k, v in data.items():
        f = interp1d((v['time'] - t_min) / 1000., v['data'])
        resampled_data[k] = f(t)
        
    # return the resampled data
    return resampled_data
```

Define a function to embed videos that have been uploaded to [Illinois Mediaspace](https://mediaspace.illinois.edu/).

```python
def mediaspace(url, width=640, height=397):
    video_id = url.rsplit('/', 1)[-1]
    src = f'https://mediaspace.illinois.edu/embed/secure/iframe/entryId/{video_id}?st=0'
    display(IFrame(src, width=width, height=height))
    display(HTML(f'<a href="{url}" target="_blank">Go to video on Illinois Mediaspace</a>'))
```

# 3. Find the force parameter

### 3.1 Move the drone up and down

```python
mediaspace('https://mediaspace.illinois.edu/media/t/1_bsk0nepq')
```

### 3.2 Get and plot flight data

Load and resample data.

```python
data = load_hardware_data(
    'force_flight_project.json',
    t_min_offset=1.,
    t_max_offset=1.,
)
```

Parse data to get:
* time
* the accelerometer measurements (in units of "g's")
* the motor power commands

```python
t = data['time']
a_x = g * data['acc.x']
a_y = g * data['acc.y']
a_z = g * data['acc.z']
m_1 = data['motor.m1']
m_2 = data['motor.m2']
m_3 = data['motor.m3']
m_4 = data['motor.m4']
```

Plot accelerometer measurements and motor power commands.

```python
fig, (ax_a, ax_m) = plt.subplots(2, 1, figsize=(12, 12), sharex=True)
ax_a.plot(t, a_x, label='a_x')
ax_a.plot(t, a_y, label='a_y')
ax_a.plot(t, a_z, label='a_z')
ax_a.legend()
ax_a.grid()
ax_m.plot(t, m_1, label='m_1')
ax_m.plot(t, m_2, label='m_2')
ax_m.plot(t, m_3, label='m_3')
ax_m.plot(t, m_4, label='m_4')
ax_m.legend()
ax_m.grid()
```

```python
t = data['time']
x = data['stateEstimate.x']
y = data['stateEstimate.y']
z = data['stateEstimate.z']

fig, ax_a = plt.subplots(1, 1, figsize=(6, 6), sharex=True)
ax_a.plot(t, x, label='x')
ax_a.plot(t, y, label='y')
ax_a.plot(t, z, label='z')
ax_a.legend()
ax_a.grid()
```

### 3.3 Apply linear regression to estimate $k_F$

First, compute `x`.

```python
x = m_1 + m_2 + m_3 + m_4 
```

Second, compute `y`.

```python
y = m * a_z
```

Third, compute `c_est`.

```python
c_est = np.sum(y * x) / np.sum(x**2) 
```

Finally, compute `k_F` in terms of `c_est`.

```python
k_F = c_est 
```

Show the estimate.

```python
print(f'k_F = {k_F:.2e}')
```

### 3.4 Validate your estimate

Compare the linear fit to the raw data.

```python
plt.figure(figsize=(10, 10))
plt.plot(x, y, '.', markersize=4, label='raw data')
plt.plot([0, 262140], [c_est * 0, c_est * 262140], label='linear fit')
plt.xlim(0, 262140)
plt.xlabel('x')
plt.ylim(0, 0.5)
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

Compare the z-axis accelerometer measurement with the prediction based on the estimate of $k_F$.

```python
a_z_predicted = x * c_est/m 

plt.figure(figsize=(12, 6))
plt.plot(t, a_z, label='a_z (measured)')
plt.plot(t, a_z_predicted, '--', label='a_z (predicted)')
plt.grid()
plt.legend()
plt.show()
```

# 4. Find the moment parameter

### 4.1 Yaw the drone back and forth

```python
mediaspace('https://mediaspace.illinois.edu/media/t/1_44ilbgfk')
```

### 4.2 Get and plot flight data

Load and resample data.

```python
data = load_hardware_data(
    'yaw_flight_project.json',
    t_min_offset=1.,
    t_max_offset=1.,
)
```

Parse data to get:
* time
* the gyroscope measurements (in units of degrees / second)
* the motor power commands

```python
t = data['time']
w_x = np.deg2rad(data['gyro.x'])
w_y = np.deg2rad(data['gyro.y'])
w_z = np.deg2rad(data['gyro.z'])
m_1 = data['motor.m1']
m_2 = data['motor.m2']
m_3 = data['motor.m3']
m_4 = data['motor.m4']
```

Find the time step. It should be `0.01` because data were sampled at 100 Hz.

```python
dt = t[1] - t[0]
print(f'dt = {dt}')
```

Plot gyroscope measurements and motor power commands.

```python
fig, (ax_w, ax_m) = plt.subplots(2, 1, figsize=(12, 12), sharex=True)
ax_w.plot(t, w_x, label='w_x')
ax_w.plot(t, w_y, label='w_y')
ax_w.plot(t, w_z, label='w_z')
ax_w.legend()
ax_w.grid()
ax_m.plot(t, m_1, label='m_1')
ax_m.plot(t, m_2, label='m_2')
ax_m.plot(t, m_3, label='m_3')
ax_m.plot(t, m_4, label='m_4')
ax_m.legend()
ax_m.grid()
```

### 4.3 Use finite difference to estimate $\dot{w}_z$

Estimate $\dot{w}_z$ by finite difference.

```python
w_z_dot = (w_z[1:] - w_z[:-1]) / dt 
```

Plot both $w_z$ and $\dot{w}_z$.

```python
plt.figure(figsize=(12, 6))
plt.plot(t, w_z, label='$w_z$')
plt.plot(t[:-1], w_z_dot, '--', label='$\dot{w}_z$')
plt.grid()
plt.legend(fontsize=16)
plt.show()
```

### 4.4 Apply linear regression to estimate $k_M$

First, compute `x`.

```python
x = - m_1 + m_2 - m_3 + m_4 
```

Second, compute `y`.

```python
y = J_z * w_z_dot
```

Third, compute `c_est`.

```python
x = x[1:]
c_est = c_est = np.sum(y * x) / np.sum(x**2)
```

Finally, compute `k_M` in terms of `c_est`.

```python
k_M = c_est 
```

Show the estimate.

```python
print(f'k_M = {k_M:.2e}')
```

### 4.4 Validate your estimate

Compare the linear fit to the raw data.

```python
plt.figure(figsize=(10, 10))
plt.plot(x, y, '.', markersize=12, label='raw data')
plt.plot([-131070, 131070], [c_est * -131070, c_est * 131070], label='linear fit')
plt.xlim(-131070, 131070)
plt.xlabel('x')
plt.ylim(-.002, 0.002)
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

Compare the finite-difference estimate of $\dot{w}_z$ to what would have been predicted using $k_M$.

```python
w_z_dot_predicted = k_M/J_z * x 

plt.figure(figsize=(12, 6))
plt.plot(t[1:], w_z_dot, label='$\dot{w}_z$ (measured)')
plt.plot(t[1:], w_z_dot_predicted, '--', label='$\dot{w}_z$ (predicted)')
plt.grid()
plt.legend()
plt.show()
```
